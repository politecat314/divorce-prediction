# **use for loop to get the average rate for each question**

```python
for column in X:
    out = X[column].mean()
    print('The mean for',column,'is:',out)
```

## **use for loop to get the varience for each question**

```python
for column in X:
    out = X[column].var()
    print('The variance for',column,'is:',out)
```
